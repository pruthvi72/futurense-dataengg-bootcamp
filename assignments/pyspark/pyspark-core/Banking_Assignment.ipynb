```python
import findspark
findspark.init()

from pyspark import SparkContext
sc = SparkContext()
```

```python
rdd = sc.textFile(r"C:\Users\Miles\futurense_hadoop-pyspark\labs\dataset\bankmarket\bankmarketdata.csv")
count = rdd.count()
print(count)
```

```python
rdd1 = rdd.map(lambda x : x.split(","))
rdd2 = rdd1.map(lambda y : y[0].split(";"))
rdd2.collect()
```

```python
rdd3 = rdd2.filter(lambda x : x[-1] == '"yes"')
count_yes = rdd3.count()
count_yes
```

```python
success_rate = count_yes/count
success_rate*100
```

```python
#2
rdd4 = rdd2.filter(lambda x : x[-1] == '"no"')
count_no = rdd4.count()
count_no
```

```python
failure_rate = count_no/count
failure_rate*100
```

```python
#3
header = rdd2.first()
rdd5 = rdd2.filter(lambda x : x!=header)
```

```python
rdd5.collect()
```

```python
rdd6 = rdd5.map(lambda x : int(x[0]))
```

```python
rdd6.max()
```

```python
rdd6.min()
```

```python
rdd6.mean()
```

```python
#4
rdd7 = rdd5.map(lambda x : int(x[5])).sortBy(lambda y: y)
```

```python
rdd7.mean()
```

```python
median = rdd7.count()//2
rdd7.collect()[median]
```

```python
#5
rdd8 = rdd3.map(lambda x : (x[0], 1))
rdd9 = rdd8.groupByKey().mapValues(list)
rdd9.map(lambda x : (x[0], len(x[1]))).collect()
```

```python
#6
rdd10 = rdd3.map(lambda x : (x[2], 1))
rdd11 = rdd10.groupByKey().mapValues(list)
rdd11.map(lambda x : (x[0], len(x[1]))).collect()
```

```python
#7
rdd12 = rdd3.map(lambda x : (x[0], x[2], 1))
rdd13 = rdd12.groupBy(lambda y : (y[0], y[1])).mapValues(list)
rdd13.map(lambda x : (x[0], len(x[1]))).collect()
```
